```python
# import cleaned csv file from Part 2
import pandas as pd

df = pd.read_csv('cleaned_data.csv')
df.head()
```

# Question 1
## What events are most effective at producing computing degree students at CCM? (What events should CCM prioritize?) 

### Quantify the number of people that participated at events at CCM prior to enrolling
### count the yes values and get a sumn for each prior_particiapation column
### Show a bar chart sorted from highest to lowest

```python
# import Matplotlib for bar graph visualization
import matplotlib.pyplot as plt 

# Use the count() function to count the number of items with a certain value
open_house_participants = df[df['prior_participation_open_house']=='Yes']['prior_participation_open_house'].count()
instant_decision_day_participants = df[df['prior_participation_instant_decision_day']=='Yes']['prior_participation_instant_decision_day'].count()
on_campus_info_session_participants = df[df['prior_participation_on_campus_info_session']=='Yes']['prior_participation_on_campus_info_session'].count()
virtual_info_session_participants = df[df['prior_participation_virtual_info_session']=='Yes']['prior_participation_virtual_info_session'].count()
women_who_dare_participants = df[df['prior_participation_women_who_dare']=='Yes']['prior_participation_women_who_dare'].count()
regional_college_fair_participants = df[df['prior_participation_regional_college_fair']=='Yes']['prior_participation_regional_college_fair'].count()
hs_sharetime_info_session_participants = df[df['prior_participation_hs_sharetime_info_session']=='Yes']['prior_participation_hs_sharetime_info_session'].count()
challenger_program_participants = df[df['prior_participation_challenger_program']=='Yes']['prior_participation_challenger_program'].count()
cybersec_program_participants = df[df['prior_participation_cybersec_program']=='Yes']['prior_participation_cybersec_program'].count()
hs_info_session_participants = df[df['prior_participation_hs_info_session']=='Yes']['prior_participation_hs_info_session'].count()
hs_campus_visit_participants = df[df['prior_participation_hs_campus_visit']=='Yes']['prior_participation_hs_campus_visit'].count()
individual_visit_participants = df[df['prior_participation_individual_campus_visit']=='Yes']['prior_participation_individual_campus_visit'].count()
workforce_dev_participants = df[df['prior_participation_workforce_development_class']=='Yes']['prior_participation_workforce_development_class'].count()
corp_training_participants = df[df['prior_participation_corporate_training']=='Yes']['prior_participation_corporate_training'].count()
teen_arts_fests_participants = df[df['prior_participation_teen_arts_fest']=='Yes']['prior_participation_teen_arts_fest'].count()
summer_camp_participants = df[df['prior_participation_CCM_summer_camp']=='Yes']['prior_participation_CCM_summer_camp'].count()

# dictionary with the data extracted above
data = {'Open House': open_house_participants , 'Instant Decision Day': instant_decision_day_participants, 'On Campus Info Session': on_campus_info_session_participants, 'Virtual Info Session': virtual_info_session_participants, 'Women Who Dare': women_who_dare_participants, 'Regional College Fair': regional_college_fair_participants, 'HS Sharetime Info Session': hs_sharetime_info_session_participants, 'Challenger Program':challenger_program_participants, 'Cybersec Program': cybersec_program_participants, 'HS Info Session': hs_info_session_participants, 'HS Campus Visit': hs_campus_visit_participants, 'Individual Campus Visit': individual_visit_participants, 'Workforce Development': workforce_dev_participants, 'Corporate Training': corp_training_participants, 'Teen Arts Fest': teen_arts_fests_participants, 'CCM Summer Camp': summer_camp_participants}

# x axis
events = list(data.keys())
# y axis
number_of = list(data.values())

# size of bar graph
fig = plt.figure(figsize = (16, 9))
# create bar graph
plt.barh(events, number_of, color = 'seagreen')
plt.ylabel("Events", fontsize = 16)
plt.xlabel("Number of students", fontsize = 16)
plt.title("Events Students Have Attended Prior to Enrolling at CCM", fontsize = 24)
plt.show() 
```

## From the data above, we can answer question 1 by confidently saying that individual campus visits are the most effective producers of computing degree students. The Open House is also an event that CCM can prioritize to improve computing degree enrollment. CCM can also advance/advertise their various info sessions to prospective students, including high schoolers.

# Question 2 
## What were the main sources from which the form's respondents received information about CCMâ€™s computing programs? (What is the most effective method to communicate with prospective students?)

```python
# Use the count() function to count the number of items with a certain value
hs_guidance_counselor = df[df['info_source_hs_guidance_counselor']=='Yes']['info_source_hs_guidance_counselor'].count()
hs_teacher = df[df['info_source_hs_teacher']== 'Yes']['info_source_hs_teacher'].count()
website = df[df['info_source_ccm_website']== 'Yes']['info_source_ccm_website'].count()
admissions_office = df[df['info_source_ccm_admissions']== 'Yes']['info_source_ccm_admissions'].count()
ccm_advisor = df[df['info_source_ccm_advisor']== 'Yes']['info_source_ccm_advisor'].count()
employer = df[df['info_source_employer']== 'Yes']['info_source_employer'].count()
workforce_dev = df[df['info_source_workforce_development']== 'Yes']['info_source_workforce_development'].count()
NJ_workforce_dev = df[df['info_source_NJ_workforce_development']== 'Yes']['info_source_NJ_workforce_development'].count()
other = df[df['info_source_other']=='Yes']['info_source_other'].count()

# dictionary with the data extracted above
data2 = {'HS Guidance Counselor': hs_guidance_counselor , 'HS Teacher': hs_teacher, 'CCM Website': website, 'CCM Admissions Office': admissions_office, 'CCM Advisor': ccm_advisor, 'Employer': employer, 'Workforce Development': workforce_dev, 'NJ Workforce Development': NJ_workforce_dev, 'Other': other}

# x axis
info_sources = list(data2.keys())
# y axis
num_of_info_sources = list(data2.values())

# size of bar graph
fig2 = plt.figure(figsize = (12, 6))
# create bar graph
plt.barh(info_sources, num_of_info_sources, color = 'pink')
plt.ylabel("Info Sources", fontsize = 16)
plt.xlabel("Number of Students Who Used Info Source", fontsize = 16)
plt.title("Students' Main Sources of Information of CCM's Computing Programs", fontsize = 24)
plt.show() 
```

## From the data above, we can answer question 2 by saying that CCM's advertisement strategies/information sources such as CCM's advisors, admissions office, and the website are very effective. 

# Question 3
## Which were the greatest motivations to seek a computing degree/certificate at CCM? (What should CCMâ€™s CS department advertise?)

```python
# Use the count() function to count the number of items with a certain value
computing_job = df[df['motivation_computing_job']=='Yes']['motivation_computing_job'].count()
transfer_bachelors = df[df['motivation_transfer_bachelors']=='Yes']['motivation_transfer_bachelors'].count()
hs_dual_enrollment = df[df['motivation_hs_dual_enrollment']=='Yes']['motivation_hs_dual_enrollment'].count()
career_advancement = df[df['motivation_career_advancement']=='Yes']['motivation_career_advancement'].count()
career_change = df[df['motivation_career_change']=='Yes']['motivation_career_change'].count()
professional_development = df[df['motivation_profession_development']=='Yes']['motivation_profession_development'].count()
job_displacement = df[df['motivation_job_displacement']=='Yes']['motivation_job_displacement'].count()
relocation = df[df['motivation_relocation']=='Yes']['motivation_relocation'].count()
keepup_tech = df[df['motivation_keep_current_in_tech']=='Yes']['motivation_keep_current_in_tech'].count()
IT_certification = df[df['motivation_IT_certification']=='Yes']['motivation_IT_certification'].count()
financial = df[df['motivation_financial']=='Yes']['motivation_financial'].count()
personal_enrichment = df[df['motivation_personal_enrichment']=='Yes']['motivation_personal_enrichment'].count()

# dictionary with the data extracted above
data3 = {'Computing Job': computing_job , 'Transfer to Bachelors': transfer_bachelors, 'HS Dual Enrollment': hs_dual_enrollment, 'Career Advancement': career_advancement, 'Career Change': career_change, 'Professional Development': professional_development, 'Job Displacment': job_displacement, 'Relocation': relocation, 'Keep Current In Tech': keepup_tech, 'IT Certification': IT_certification, 'Financial': financial, 'Personal Enrichment': personal_enrichment}

# x axis
motivations = list(data3.keys())
# y axis
num_motivations = list(data3.values())

# size of bar graph
fig2 = plt.figure(figsize = (12, 8))
# create bar graph
plt.barh(motivations, num_motivations, color = 'lightblue')
plt.ylabel("Motivations", fontsize = 16)
plt.xlabel("Number of Students with Motivation", fontsize = 16)
plt.title("Students' Main Motivations for Enrolling in CCM's Computing Programs", fontsize = 24)
plt.show() 
```

## From the data above, we observe that students want to enroll in computing programs for a career with computing (job or bachelors degree). To motativate people to enroll in computing degrees, CCM can advertise their prestigious IT certification and their low tuition fees.
